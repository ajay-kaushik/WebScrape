# Final Code

```python
# import require libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# t20 series url 
t20_URL = 'https://stats.espncricinfo.com/wi/content/records/335433.html'

#base url
base_URL = 'https://stats.espncricinfo.com'

def get_soup(url):
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html.parser') 
    return soup

def get_series_url_list(series_soup):
    selection_class = 'data-link'
    tag_content  = series_soup.find_all('a',{'class':selection_class})
    
    sub_str = '/wi/engine/series/'
    url_list = []
    for tag in tag_content:
        
        if sub_str in tag['href']:
            #print('found:',tag['href'])
            url_list.append(base_URL + tag['href'])
            
    return url_list

def save_as_csv(fileName, dict_file):
    df = pd.DataFrame(dict_file)
    df.to_csv(fileName +'.csv', index=None)
    print("saved..!!")
    #retrun dataframe
    return  df

def get_match_url_list(series_list):
    selection_class = 'widget-tab-link'
    sub_str = "match-results"

    selection_class_match = 'match-info-link-FIXTURES'
    final_matches_list = []

    for series in series_list:
        print("Extracting...",series)
        # call get_soup() function
        series_soup = get_soup(series)
        matches_content = series_soup.find_all('a',{'class':selection_class})

        for matches in matches_content:
            if sub_str in str(matches):
                match_soup = get_soup(base_URL+ matches['href'])
                match_content = match_soup.find_all('a',{'class':selection_class_match})
                
                for match in match_content:
                    final_matches_list.append(base_URL+match['href'])
                    #print( "Found the URL:", base_URL + match['href'])

    print('Extraction completed...!!!')
    return final_matches_list
```

```python
# call the function

series_soup = get_soup(t20_URL)
series_url_list = get_series_url_list(series_soup)

#use dict for column  name
series_dict = {
    'series_url':series_url_list
}
filename = "t20_series_url"
df = save_as_csv(filename, series_dict)


match_url = get_match_url_list(series_url_list)

match_dict = {
    'match_url':match_url
}
filename = "t20_match_url"
match_df = save_as_csv(filename, match_dict)
```
